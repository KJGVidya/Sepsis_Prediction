<h4>Finding most correlated columns

```python
def correlation_matrix(df_corr,threshold):
    df_lower=df_corr.where(np.tril(np.ones(corr.shape),k=-1).astype(bool))
    df_lower=df_lower[(df_lower>threshold) | (df_lower<-threshold)]
    df_lower.dropna(how='all',axis=1,inplace=True)
    df_lower.dropna(how='all',axis=0,inplace=True)
    return df_lower
```

```python
correlation_matrix(corr,0.9)
```

Unit1 and Unit2 are negatively correlated.

Hgb snd Hct, ICULOS and Hour, Bilirubin_total and Bilirubin_Direct are positively correlated.

<h4>Vital Signs Time series Analysis - Visualizing onset of sepsis for random patients

```python
import random
from random import randrange
def SepRandom_sampling(sep_data,n=35):
    """Truncating sepsis patient record who have more than 35 records to anywhere between 20 to 35"""
    dict_pat=sep_data.groupby(['Patient_ID']).size().to_dict()
    patient_list=[(k) for k,v in dict_pat.items() if v>35]
    random.shuffle(patient_list)
    pat_split = np.array_split(patient_list,n)
    sep_random=[]
    for i in range(0,n):
        sep_random.append(sep_data[sep_data.Patient_ID.isin(pat_split[i])].groupby('Patient_ID').tail(randrange(20,35)))
    data1=pd.concat(sep_random)
    data1['Hour']=data1.groupby(by=['Patient_ID'])['Hour'].apply(lambda x: x.rank(method='min'))
    data2=sep_data[~sep_data.Patient_ID.isin(patient_list)]
    data = pd.concat([data1,data2],axis=0)
    return data
```

```python
"""Forward and Backward filling as most of the data is missing"""
grouped_by_patient = df_sep.groupby('Patient_ID')
df_sep = grouped_by_patient.apply(lambda x: x.bfill().ffill())
```

```python
"""Makes sure we have atleast 10 data points before sepsis onset to visualize variations before and after sepsis onset"""
sep_pat1= df_sep[(df_sep['SepsisLabel']==0)& (df_sep['Hour']==11)].Patient_ID.unique()
pat_rand=random.sample(list(sep_pat1), 10)
df_1=df_sep[df_sep.Patient_ID.isin(pat_rand)]
'''Sampling only last few rows for visualization'''
df_2=SepRandom_sampling(df_1)
```
